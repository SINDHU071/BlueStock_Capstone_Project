```python
schemes = {
    "SBI_Bluechip": 119551,
    "ICICI_Bluechip": 120503,
    "Nippon_LargeCap": 118632,
    "Axis_Bluechip": 119092,
    "Kotak_Bluechip": 120841
}
```

```python
for name, code in schemes.items():

    url = f"https://api.mfapi.in/mf/{code}"

    response = requests.get(url)

    data = response.json()

    df = pd.DataFrame(data['data'])

    save_file = f"/content/drive/MyDrive/MutualFundAnalytics/Data/Raw/{name}.csv"

    df.to_csv(save_file, index=False)

    print(f"{name} saved successfully")
```

```python
nav_history = pd.read_csv(
    f"{raw_path}/Copy of 02_nav_history.csv"
)

nav_history.head()
```

```python
nav_history.columns.tolist()
```

```python
missing_codes = set(
    fund_master['amfi_code']
) - set(
    nav_history['amfi_code']
)

print("Missing Codes:", len(missing_codes))
```

```python
for col in fund_master.columns:
    print(col, ":", fund_master[col].isnull().sum())
```

```python
print(
    "Duplicate Rows:",
    fund_master.duplicated().sum()
)
```

```python
quality_report = f"""
DATA QUALITY REPORT

Fund Master Records:
{len(fund_master)}

NAV History Records:
{len(nav_history)}

Duplicate Records:
{fund_master.duplicated().sum()}

Missing AMFI Codes:
{len(missing_codes)}
"""

print(quality_report)
```

```python
with open(
"/content/drive/MyDrive/MutualFundAnalytics/Reports/data_quality_report.txt",
"w"
) as f:

    f.write(quality_report)

print("Data Quality Report Saved")
```
